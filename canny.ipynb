```python
import cv2
from diffusers.utils import load_image
from pathlib import Path
import numpy as np
from PIL import Image


image_path = Path("media/inputs/spiral.png")
low = 200
high = 450

image = load_image(str(image_path))
canny = cv2.Canny(np.array(image), low, high)
canny = np.concatenate([canny[:, :, None]] * 3, axis=2)
canny_img = Image.fromarray(canny)
save_path = Path("media/inputs/canny") / image_path.name
canny_img.save(save_path)
canny_img
```
